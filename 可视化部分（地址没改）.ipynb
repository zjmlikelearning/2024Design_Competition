(4) 随着乡村人口的减少和就业人员数量的下降，农村经济结构可能会发生变化，转向非农产业或服务业，这也会影响对耕地的利用方式和需求。

### **1.2.1 异常值处理**

```python
# 检测 '已流转承包地面积' 数据值比 '承包地面积' 数据值大的异常行并输出
abnormal_rows = Land_basic_conditions[Land_basic_conditions['已流转承包地面积'] > Land_basic_conditions['承包地面积']]
abnormal_rows[['承包地面积', '已流转承包地面积']]
```

使用**3-Sigma方法**检测各列异常值。具体步骤如下：  
(1) 计算各列的平均值和标准差。  
(2) 根据3-Sigma准则，计算异常值的上下界，即平均值加减3倍标准差。  
(3) 将超出上下界范围的值标记为异常值，并将结果存储在新的列中，异常值用1表示，非异常值用0表示。

```python
Village_area_mean = np.mean(Land_basic_conditions['本村占地总面积']) # 本村占地总面积平均值
Village_area_std = np.std(Land_basic_conditions['本村占地总面积'])   # 本村占地总面积标准差

# 3-Sigma准则
Village_area_low_bound = Village_area_mean - 3 * Village_area_std
Village_area_upper_bound = Village_area_mean + 3 * Village_area_std
print(f'Village_area low bound: {Village_area_low_bound}, Village_area upper bound: {Village_area_upper_bound}')

# 判定异常值数量
Land_basic_conditions['Village_area_three_sigma'] = 0
Land_basic_conditions.loc[(Land_basic_conditions['本村占地总面积'] > Village_area_upper_bound) | (Land_basic_conditions['本村占地总面积'] < Village_area_low_bound), 'Village_area_three_sigma'] = 1
Outliers_num = len(Land_basic_conditions[Land_basic_conditions['Village_area_three_sigma'] == 1])
if Outliers_num != 0:
    print('3-Sigma方法检测出的本村占地总面积异常值数量:{}'.format(Outliers_num))
else:
    print('未检测出异常值！')
```

```python
Cultivated_area_mean = np.mean(Land_basic_conditions['本村耕地总面积']) # 本村耕地总面积平均值
Cultivated_area_std = np.std(Land_basic_conditions['本村耕地总面积'])   # 本村耕地总面积标准差

# 3-Sigma准则
Cultivated_area_low_bound = Cultivated_area_mean - 3 * Cultivated_area_std
Cultivated_area_upper_bound = Cultivated_area_mean + 3 * Cultivated_area_std
print(f'Cultivated_area low bound: {Cultivated_area_low_bound}, Cultivated_area upper bound: {Cultivated_area_upper_bound}')

# 判定异常值数量
Land_basic_conditions['Cultivated_area_three_sigma'] = 0
Land_basic_conditions.loc[(Land_basic_conditions['本村耕地总面积'] > Cultivated_area_upper_bound) | (Land_basic_conditions['本村耕地总面积'] < Cultivated_area_low_bound), 'Cultivated_area_three_sigma'] = 1
Outliers_num = len(Land_basic_conditions[Land_basic_conditions['Cultivated_area_three_sigma']==1])
if Outliers_num != 0:
    print('3-Sigma方法检测出的本村耕地总面积异常值数量:{}'.format(Outliers_num))
else:
    print('未检测出异常值！')
```

```python
Contracted_area_mean = np.mean(Land_basic_conditions['承包地面积']) # 承包地面积平均值
Contracted_area_std = np.std(Land_basic_conditions['承包地面积'])   # 承包地面积标准差

# 3-Sigma准则
Contracted_area_low_bound = Contracted_area_mean - 3 * Contracted_area_std
Contracted_area_upper_bound = Contracted_area_mean + 3 * Contracted_area_std
print(f'Contracted_area low bound: {Contracted_area_low_bound}, Contracted_area upper bound: {Contracted_area_upper_bound}')

# 判定异常值数量
Land_basic_conditions['Contracted_area_three_sigma'] = 0
Land_basic_conditions.loc[(Land_basic_conditions['承包地面积'] > Contracted_area_upper_bound) | (Land_basic_conditions['承包地面积'] < Contracted_area_low_bound), 'Contracted_area_three_sigma'] = 1
Outliers_num = len(Land_basic_conditions[Land_basic_conditions['Contracted_area_three_sigma']==1])
if Outliers_num != 0:
    print('3-Sigma方法检测出的承包地面积异常值数量:{}'.format(Outliers_num))
else:
    print('未检测出异常值！')
```

```python
Transferred_area_mean = np.mean(Land_basic_conditions['已流转承包地面积']) # 已流转承包地面积平均值
Transferred_area_std = np.std(Land_basic_conditions['已流转承包地面积'])   # 已流转承包地面积标准差

# 3-Sigma准则
Transferred_area_low_bound = Transferred_area_mean - 3 * Transferred_area_std
Transferred_area_upper_bound = Transferred_area_mean + 3 * Transferred_area_std
print(f'Transferred_area low bound: {Transferred_area_low_bound}, Transferred_area upper bound: {Transferred_area_upper_bound}')

# 判定异常值数量
Land_basic_conditions['Transferred_area_three_sigma'] = 0
Land_basic_conditions.loc[(Land_basic_conditions['已流转承包地面积'] > Transferred_area_upper_bound) | (Land_basic_conditions['已流转承包地面积'] < Transferred_area_low_bound), 'Transferred_area_three_sigma'] = 1
Outliers_num = len(Land_basic_conditions[Land_basic_conditions['Transferred_area_three_sigma']==1])
if Outliers_num != 0:
    print('3-Sigma方法检测出的已流转承包地面积异常值数量:{}'.format(Outliers_num))
else:
    print('未检测出异常值！')
```

```python
# 使用本村占地总面积的均值填分析数据充异常值
Land_basic_conditions.loc[(Land_basic_conditions['本村占地总面积'] > Village_area_upper_bound) | (Land_basic_conditions['本村占地总面积'] < Village_area_low_bound), '本村占地总面积'] = Village_area_mean

# 使用本村耕地总面积的均值填充分析数据异常值
Land_basic_conditions.loc[(Land_basic_conditions['本村耕地总面积'] > Cultivated_area_upper_bound) | (Land_basic_conditions['本村耕地总面积'] < Cultivated_area_low_bound), '本村耕地总面积'] = Cultivated_area_mean

# 使用承包地面积的均值填充分析数据异常值
Land_basic_conditions.loc[(Land_basic_conditions['承包地面积'] > Contracted_area_upper_bound) | (Land_basic_conditions['本村耕地总面积'] < Contracted_area_low_bound), '承包地面积'] = Contracted_area_mean

# 使用已流转承包地面积的均值填充分析数据异常值
Land_basic_conditions.loc[(Land_basic_conditions['已流转承包地面积'] > Transferred_area_upper_bound) | (Land_basic_conditions['本村耕地总面积'] < Transferred_area_low_bound), '已流转承包地面积'] = Transferred_area_mean
```

```python
# 删除多余列
Land_basic_conditions = Land_basic_conditions.drop(['Village_area_three_sigma', 'Cultivated_area_three_sigma', 'Contracted_area_three_sigma', 'Transferred_area_three_sigma'], axis=1)
# 重置索引
Land_basic_conditions.reset_index(drop = True, inplace = True)

Land_basic_conditions.head()
```

### **1.2.5 土地超耕问题**

```python
# 根据调查年份字段分类统计每年的超耕率和超耕面积

# 使用 groupby 函数按照调查年份字段对数据进行分组
grouped = Land_basic_conditions.groupby('调查年份')

# 创建空列表用于存储每年的数据
years = []             # 存储年份
overplowed_areas = []  # 存储每年的超耕面积
overplow_rates = []    # 存储每年的超耕率

# 遍历分组后的数据
for year, group in grouped:
    # 将当前年份添加到年份列表中
    years.append(year)
    
    # 计算当前年份中超耕的行（本村耕地总面积大于本村占地总面积）
    overplowed_rows = group[group['本村耕地总面积'] > group['本村占地总面积']]
    
    # 计算超耕的数量和超耕的面积
    overplowed_count = len(overplowed_rows)
    overplowed_area = overplowed_rows['本村耕地总面积'].sum() - overplowed_rows['本村占地总面积'].sum()
    overplowed_areas.append(overplowed_area)
    
    # 计算超耕率（超耕的数量占总样本数量的比例）
    total_rows = len(group)
    if total_rows > 0:
        overplow_rate = overplowed_count / total_rows
    else:
        overplow_rate = 0
    overplow_rate = overplow_rate * 100  # 转换为百分比
    
    # 将超耕率添加到列表中
    overplow_rates.append(overplow_rate)
```

**绘制柱状-折线组合图以展示每年的超耕面积和超耕率变化情况**

```python
overplow_data = {
    '年份': years,
    '超耕面积': overplowed_areas,
    '超耕率': overplow_rates
}

overplow_data = pd.DataFrame(overplow_data)

overplow_data['超耕率'] = [round(x, 2) for x in overplow_data['超耕率']]

bar = (
    Bar(init_opts=opts.InitOpts(width='960px', height='540px', theme=ThemeType.VINTAGE))
    .add_xaxis(overplow_data['年份'].astype(str).tolist())
    .add_yaxis(
        "超耕面积(亩)",
        overplow_data['超耕面积'].tolist(),
        gap="0%",
        category_gap="50%",
        itemstyle_opts=opts.ItemStyleOpts(
            color={
                "type": 'linear',
                "x": 0,
                "y": 0,
                "x2": 0,
                "y2": 1,
                "colorStops": [{
                    "offset": 0, "color": '#ABD882' # 渐变色填充柱状图
                }, {
                    "offset": 1, "color": '#3B7658' 
                }],
            }
        )
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .extend_axis(
        yaxis=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(formatter="{value} %", color="rgb(22, 22, 23)"), interval=5
        )
    )
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-45, color="rgb(22, 22, 23)")),  # 设置 x 轴标签旋转角度
        yaxis_opts=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(formatter="{value} 亩", color="rgb(22, 22, 23)"),  # 设置 y 轴标签格式
        ),
        visualmap_opts=opts.VisualMapOpts(),  # 设置视觉映射组件
        datazoom_opts=[opts.DataZoomOpts(type_="inside")],  # 设置数据缩放组件
    )
    .set_global_opts(datazoom_opts=[opts.DataZoomOpts(type_="slider", range_start=0, range_end=100, orient="horizontal", pos_bottom="-0.5%")])  # 设置数据缩放滑块组件
    .set_global_opts(legend_opts=opts.LegendOpts(orient="horizontal", pos_left="center", pos_top="5%", textstyle_opts=opts.TextStyleOpts(font_size=18)))  # 设置图例位置和方向
)

line = (
    Line()
    .add_xaxis(overplow_data['年份'].astype(str).tolist())
    .add_yaxis(
        "超耕率",
        overplow_data['超耕率'].tolist(),
        yaxis_index=1,
        label_opts=opts.LabelOpts(is_show=True),
        linestyle_opts=opts.LineStyleOpts(width=4, type_='dashed'),
    )
    .set_series_opts(itemstyle_opts=opts.ItemStyleOpts(color="rgb(250, 194, 51)"))
)

bar.overlap(line).render_notebook()
```

### **1.3.1 异常值处理**

对于存在异常值的行，使用**放缩法**替换异常值。具体操作是，对于异常值所在行，根据各类经营主体的耕地经营面积之和计算出总面积，然后按比例调整每个经营主体的面积，使其总和等于 '本村耕地总面积'。其余使用放缩法处理异常值的原理同上，即通过**按比例调整异常值以确保数据的总和不变**。

```python
# 找到各类经营主体(按类型分)耕地经营面积的和大于 '本村耕地总面积' 的异常值所在行
abnormal_values = Village_entities_type[Village_entities_type['耕地总面积中：农户家庭经营面积'] +  
Village_entities_type['耕地总面积中：村集体经营面积'] +  Village_entities_type['耕地总面积中：企业经营面积'] +  
Village_entities_type['耕地总面积中：农民专业合作经济组织经营面积'] +  
Village_entities_type['耕地总面积中：其他主体经营面积'] > Village_entities_type['本村耕地总面积']]

if not abnormal_values.empty:
    print("存在异常值的行:")
    print(abnormal_values.index)
    # 使用放缩法替换异常值
    for index in abnormal_values.index:
        Manage_sum = Village_entities_type.iloc[index, 1:6].sum()
        Village_entities_type.iloc[index, 1] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 1] / Manage_sum )
        Village_entities_type.iloc[index, 2] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 2] / Manage_sum )
        Village_entities_type.iloc[index, 3] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 3] / Manage_sum )
        Village_entities_type.iloc[index, 4] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 4] / Manage_sum )
        Village_entities_type.iloc[index, 5] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 5] / Manage_sum )
        
else:
    print("未发现异常值")
```

```python
# 找到各类经营主体(按经营面积分)耕地经营面积的和大于 '本村耕地总面积' 的异常值所在行
abnormal_values = Village_entities_type[Village_entities_type['耕地总面积中：由经营面积超过500亩的主体经营'] +  
Village_entities_type['耕地总面积中：由经营面积介于200-500亩的主体经营'] +  Village_entities_type['耕地总面积中：由经营面积介于100-200亩的主体经营'] +  
Village_entities_type['耕地总面积中：由经营面积介于50-100亩的主体经营'] +  
Village_entities_type['耕地总面积中：由村外经营主体经营'] > Village_entities_type['本村耕地总面积']]

if not abnormal_values.empty:
    print("存在异常值的行:")
    print(abnormal_values.index)
    # 使用放缩法替换异常值
    for index in abnormal_values.index:
        Manage_sum = Village_entities_type.iloc[index, 6:11].sum()
        Village_entities_type.iloc[index, 6] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 6] / Manage_sum )
        Village_entities_type.iloc[index, 7] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 7] / Manage_sum )
        Village_entities_type.iloc[index, 8] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 8] / Manage_sum )
        Village_entities_type.iloc[index, 9] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 9] / Manage_sum )
        Village_entities_type.iloc[index, 10] = Village_entities_type.iloc[index, 0] * ( Village_entities_type.iloc[index, 10] / Manage_sum )
        
else:
    print("未发现异常值")
```

**绘制动态漏斗图以展示从2020年到2022年不同年份的耕地转出过程规范化情况，包括不同转化流程的转化率信息。**

```python
# 创建动态漏斗图

def timeline_funnel() -> Timeline:
    tl = Timeline(init_opts=opts.InitOpts(width='960px', height='540px', theme=ThemeType.WONDERLAND))
    
    for i in range(2020, 2023):
        # 计算各流程转化率
        Conversion_rate = {
            "耕地转出在外的面积": round(data[i-2020][1] / data[i-2020][1], 4) * 100,
            "签订书面合同的转出面积率": round(data[i-2020][9] / data[i-2020][1], 4) * 100,
            "讲明流转期限的转出面积率": round(data[i-2020][10] / data[i-2020][1], 4) * 100,
            }
        funnel = (
            Funnel()
            .add("占比率",  # 系列名称为空字符串
                 data_pair = list(Conversion_rate.items()),
                 gap=5,
                 label_opts=opts.LabelOpts(position="inside", font_size=18, formatter = "{b}: {c}%"),
                 itemstyle_opts=opts.ItemStyleOpts(border_width=3))
            .set_global_opts(title_opts=opts.TitleOpts("{}年耕地转出过程规范化情况(亩)".format(i), pos_left="center"),
            legend_opts=opts.LegendOpts(is_show=False),
            tooltip_opts = opts.TooltipOpts(trigger = "item", formatter = "{a} <br/>{b}: {c}%")
            )
        )
        tl.add(funnel, "{}年".format(i))

    #设置自动播放
    tl.add_schema(
        play_interval = 2000,
        is_timeline_show = True,
        is_auto_play = True,
        is_loop_play = True
        )
    
    return tl

timeline_funnel().render_notebook()
```

```python
Rent_affecting_factors = Transferred_land_rent[['耕地年租金总额', '流转对象是否为3代以内近亲', '是否通过土地交易平台或中心流转']].copy()
Rent_affecting_factors.head()
```

```python
Rent_affecting_relative = Rent_affecting_factors[['耕地年租金总额', '流转对象是否为3代以内近亲']].copy()

# 删除 '流转对象是否为3代以内近亲' 列有缺失值的行
Rent_affecting_relative.dropna(subset=['流转对象是否为3代以内近亲'], inplace=True)

Rent_affecting_relative.head()
```

```python
# 计算不同流转对象下耕地年租金平均额
Relative_result = Rent_affecting_relative.groupby('流转对象是否为3代以内近亲')[['耕地年租金总额']].mean()
Relative_result
```

分析：  
观察数据可知，流转对象不是3代以内近亲的流转耕地平均年租金额为3314.92元/亩，流转对象是3代以内近亲的流转耕地平均年租金额为3873.42元/亩，后者是前者的约1.17倍；而未通过土地交易平台或中心的流转耕地平均年租金额为4140.33元/亩，通过土地交易平台或中心的流转耕地平均年租金额为3970.54元/亩，前者约是后者的1.04倍。  
可见，**流转对象是3代以内近亲的耕地平均年租金额明显高于不是3代以内近亲的情况**，可能原因如下：  
(1) 在中国传统文化中，土地往往被视为家族的重要财富，传承至亲属之间可能更受重视。  
(2) 在家庭内部或亲戚之间进行耕地流转时，双方可能更倾向于信任对方，合作意愿更高。  
(3) 家族成员间的耕地流转可能会更容易进行资源整合，实现规模化经营。  
然而，**通过土地交易平台或中心进行流转的耕地年租金额与未通过的情况相差不大**，可能原因如下：  
(1) 土地交易平台或中心是政府或相关部门建立的用于土地流转交易的机构，其存在的目的是为了规范土地流转市场，提高交易的透明度和效率，保障农民在土地流转过程中的合法权益，而不是为了提高耕地流转租金。  
(2) 一些地区可能存在相似的土地利用需求和市场情况，导致通过平台流转和未通过平台流转的耕地年租金额相差不大。例如，如果一个地区的土地利用需求主要集中在某种农业种植或养殖业，不论通过平台还是其他途径，流转的价格可能会趋于相似。  
(3) 土地交易平台或中心可能会从流转租金中扣除适量手续费。  

综上，可以针对耕地流转提出以下建议：  
(1) 政府可以通过政策扶持或宣传教育，鼓励家族成员之间的合作，促进土地资源的有效利用和管理。  
(2) 政府可以加大对土地交易平台或中心的投入和支持，提升其技术水平和服务能力，提高市场的运行效率和透明度，从而凭借优质的服务和切实的收益保障吸引更多参与者。  
(3) 相关监管部门需建立健全流转市场监管机制，加强对流转行为的监督和管理，打击不法行为，维护市场秩序。立法部门则要完善相关法律法规，明确流转交易的合规要求和责任义务，保护交易双方的合法权益，降低交易风险。
